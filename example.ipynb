# Пример использования датасета Росстата для обогащения данных ФНС

Импорт и загрузка данных:

```python
import pandas  as pd

fns = pd.read_csv('clear_fns.csv',parse_dates=True,index_col='date',)
rosstat = pd.read_csv('rosstat/week.csv',delimiter=';')
rosstat_product = pd.read_csv('rosstat/price_product.csv',delimiter=';')
demography = pd.read_csv('demography/public_demogr.csv',delimiter=';')
```

```python
fns
```

Для примера дополним данными Росстата данные ФНС по молочной продукции. Судя по списку товаров Росстата, наиболее полной категорией включающей все товары молочной группы является "Молоко и молочная продукция".

```python
display(pd.DataFrame(rosstat_product[rosstat_product['parent_name'].str.lower().str.contains('молоко')].value_counts()))
product_ids = rosstat_product[rosstat_product['parent_id']=='1100']['child_id'].values
```

Выбираем все данные о ценах продуктов из категории "Молоко и молочная продукция":

```python
rosstat_milk = rosstat[rosstat['productcode'].isin(product_ids)]
rosstat_milk = rosstat_milk[['year','week','okato','avgprice']].groupby(by=['year','week','okato']).mean()
rosstat_milk.rename({'avgprice':'rosstat_milk_kefir'},inplace=True,axis=1)
rosstat_milk.reset_index(inplace=True)
rosstat_milk.head()
```

Выбираем данные о численности населения регионов

```python
demography = demography[(demography['age_group']==83) &
           (demography['indicator']==54) &
           (demography['area_type']==1) &
           (demography['sex']==3) & 
           (demography['year']>2018)
          ].groupby(['oktmo','year']).sum().reset_index()

demography = demography[demography.oktmo.apply(lambda x: x.isnumeric())]
demography['oktmo'] = demography['oktmo'].astype('int')

demography.head()
```

Готовим данные ФНС к дополнению и объединяем:

```python
fns_milk = fns[['okato','oktmo','milk_kefir','milk_kefir_value']].copy()
fns_milk['year'] = fns_milk.index.isocalendar().year
fns_milk['week'] = fns_milk.index.isocalendar().week
fns_milk.reset_index(inplace=True)
fns_milk = fns_milk.merge(rosstat_milk, how='left', left_on=['year','week','okato'], right_on=['year','week','okato'])
fns_milk = fns_milk.merge(demography[['year','oktmo','indicator_value']], how='left', left_on=['year','oktmo'], right_on=['year','oktmo'])
fns_milk.rename({'indicator_value':'citizens'},axis=1,inplace=True)

fns_milk['milk_kefir_value_citizens'] = fns_milk['citizens']/fns_milk['milk_kefir_value']

fns_milk.head(5)
```

Теперь мы можем использовать статистические данные Росстата для обучения моделей.
